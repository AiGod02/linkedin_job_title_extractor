```python
prompt_template = PromptTemplate(
    input_variables=["concatenated_description"],
    template="""
    Extract the most senior job title from the following combined job descriptions:
    {concatenated_description}
    Please return the highest job title and the corresponding LinkedIn ID.
    """
)
```

```python
rag_chain = prompt_template | llm | StrOutputParser()
```

```python
def process_linkedin_descriptions(descriptions):
    valid_results = []
    rejected_descriptions = []

    # Concatenate descriptions into a single string
    concatenated_description = ""
    for description in descriptions:
        linkedin_id = description.get("linkedin_id", "")
        job_description = description.get("job_description", "")
        concatenated_description += f"LinkedIn ID: {linkedin_id}, Description: {job_description} | "

    # Strip the last separator " | "
    concatenated_description = concatenated_description.strip(" | ")

    # Run the LLM chain to get the highest job title and corresponding LinkedIn ID
    result = rag_chain.invoke({"concatenated_description": concatenated_description})

    if result:
        # Assume result is in format "Job Title: <title>, LinkedIn ID: <id>"
        title, linkedin_id = result.split(", LinkedIn ID: ")
        valid_results.append({
            "job_title": title.split("Job Title: ")[-1].strip(),
            "linkedin_id": linkedin_id.strip()
        })
    else:
        rejected_descriptions.append(descriptions)

    return valid_results, rejected_descriptions
```

```python

# Example LinkedIn descriptions
linkedin_descriptions = [
    {
        "linkedin_id": "12345",
        "job_description": "Lead AI engineer responsible for innovative projects..."
    },
    {
        "linkedin_id": "67890",
        "job_description": "Senior software developer working on enterprise solutions..."
    },
    {
        "linkedin_id": "54321",
        "job_description": "Junior developer with experience in web technologies..."
    }
]

# Process the descriptions
valid_results, rejected_descriptions = process_linkedin_descriptions(linkedin_descriptions)

# Print results
print("Valid Results:", valid_results)
print("Rejected Descriptions:", rejected_descriptions)
```

```python
def process_csv(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)

    # Initialize an empty string for concatenation
    combined_description = ""

    # Concatenate all job descriptions together
    for index, row in df.iterrows():
        company_name = row['company_domain_name']
        job_desc = f"Job: {row['job']}, Description: {row['desc']}, Company: {company_name}"
        combined_description += job_desc + " | "

    # Remove the last separator " | "
    combined_description = combined_description.strip(" | ")

    # Run the LLM chain to get the highest job title and corresponding company name
    result = rag_chain.invoke({"combined_description": combined_description})

    if result:
        # Assume result is in format "Job Title: <title>, Company: <company>"
        job_title, company_name = result.split(", Company: ")
        valid_result = {
            "job_title": job_title.split("Job Title: ")[-1].strip(),
            "company_name": company_name.strip()
        }
        return valid_result
    else:
        return None

# Example usage
file_path = '/content/linkedin_sample (2).csv'  # Replace with the actual path to your CSV file
valid_result = process_csv(file_path)

# Print results
if valid_result:
    print("Most Senior Job Title and Company:", valid_result)
else:
    print("No valid result found.")
```

```python
def process_csv(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)

    valid_results = []
    rejected_rows = []

    # Process each row
    for index, row in df.iterrows():
        company_name = row['company_domain_name']
        combined_description = f"Job: {row['job']}, Description: {row['desc']}"

        # Run the LLM chain to get the highest job title and corresponding company name
        result = rag_chain.invoke({"combined_description": combined_description})

        if result:
            # Assume result is in format "Job Title: <title>, Company: <company>"
            job_title = result.split(", Company: ")[0].split("Job Title: ")[-1].strip()
            valid_results.append({
                "job_title": job_title,
                "company_name": company_name
            })
        else:
            rejected_rows.append(row)

    return valid_results, rejected_rows


```
